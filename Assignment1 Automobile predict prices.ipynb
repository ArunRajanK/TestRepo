### <span style='color:DarkBlue'>1.2 Import library</span>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### <span style='color:DarkBlue'>1.3 Get the version of the pycaret</span>
- Pycaret is an open source automated library for Machine Learning. 
- It is used to compare multiple ML models (18) in just three lines of code
- It contain 55 dataset of different domain 

```python
from pycaret.utils import version
version()
```

### <span style='color:DarkBlue'>1.4 Datasets in pycaret</span>

```python
from pycaret.datasets import get_data
dataSets = get_data('index')
```

### <span style='color:DarkBlue'>1.5 Loading the dataset from pycaret</span>

```python
# Internet connection is required
boston_df = get_data("boston")

# Read data from file
# import pandas as pd
# boston_df = read.csv("myFile.csv")
# This is regression dataset. The values in medv are continuous values
```

#### Get the dimention of dataset

```python
print(boston_df.shape)
```

### <span style='color:DarkBlue'>1.6 Divide dataset into train and test</span>

```python
X = boston_df.iloc[:,0:13].values
y = boston_df.iloc[:,13].values

print("X.shape ->", X.shape)
print("y.shape ->", y.shape)
```

### <span style='color:DarkBlue'>1.7 Split dataset into train and test</span>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

print("X_train.shape ->", X_train.shape)
print("y_train.shape ->", y_train.shape)

print("\nX_test.shape  ->", X_test.shape)
print("y_test.shape  ->", y_test.shape)
```

### <span style='color:DarkBlue'>1.8 Train the linear model</span>

```python
from sklearn.linear_model import LinearRegression

linearModel = LinearRegression().fit(X_train, y_train)

# No output
```

### <span style='color:DarkBlue'>1.9 Get the coefficient (weights) of the linear model</span>

```python
linearModel.coef_
```

### <span style='color:DarkBlue'>1.10 Test the model (Generate Predicted)</span>

```python
y_predict = linearModel.predict(X_test)
y_predict
```

### <span style='color:DarkBlue'>1.11 Performance Metrics</span>

```python
from sklearn import metrics

print ("Mean Absolute Error     -->", metrics.mean_absolute_error(y_test,y_predict))
print ("Mean Squared Error      -->", metrics.mean_squared_error(y_test,y_predict))
print ("Root Mean Squared Error -->", np.sqrt(metrics.mean_squared_error(y_test,y_predict)))

cor = np.corrcoef(y_test,y_predict)[1,0]
print ("Correlation             -->", cor)
print ("RSquare                 -->", cor * cor)
```

### <span style='color:DarkBlue'>1.12 Plot and save scatter plot</span>

```python
plt.scatter(y_test, y_predict)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Actul Vs Predicted using Linear Model')
plt.savefig("result-LinearModel-ScatterPlot.jpg", dpi=300)
plt.show()
```

### <span style='color:DarkBlue'>1.13 Save actual/predicted to file</span>

```python
df_actual_predicted = pd.concat([pd.Series(y_test), pd.Series(y_predict)], axis=1)
print(df_actual_predicted.head())
df_actual_predicted.to_csv("result-LinearModel-ActualPredicted.csv")
```

### <span style='color:DarkBlue'>Explore</span>
- result-LinearModel-ScatterPlot.jpg
- result-LinearModel-ActualPredicted.csv
